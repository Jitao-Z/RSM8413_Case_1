### 1.2 Initial Cleanings

```python
# Rename the response column and format all other variable names
ebay = ebay_raw.rename(columns={'Competitive?': 'Competitive',
                                'sellerRating': 'SellerRating',
                                'endDay': 'EndDay'})
ebay.head()
```

```python
# convert all prices into USD (preparation)
# https://www.xe.com/currencytables/?from=GBP&date=2004-12-31
# FX on May 31, 2004
eur_to_usd = 1.2076
gbp_to_usd = 1.8080

def convert_fx(row):
    if row['Currency'] == 'EUR':
        row['ClosePrice'] *= eur_to_usd
        row['OpenPrice'] *= eur_to_usd
    elif row['Currency'] == 'GBP':
        row['ClosePrice'] *= gbp_to_usd
        row['OpenPrice'] *= gbp_to_usd
    return row
```

```python
# convert all prices into USD (continue) & get rid of the currency column
ebay = ebay.apply(convert_fx, axis = 1)
ebay['ClosePrice'] = ebay['ClosePrice'].round(2)
ebay['OpenPrice'] = ebay['OpenPrice'].round(2)

ebay = ebay.drop(columns=['Currency'])
```

## 5. Decision Tree – Practical Predictors

Model with **pre-auction predictors only** (OpenPrice, SellerRating, Duration, Category, EndDay).  
This setup simulates a realistic seller’s perspective, since ClosePrice is not available before the auction ends.  


```python
from sklearn.model_selection import train_test_split


train_df_prac, test_df_prac = train_test_split(
    ebay, test_size=0.3, random_state=8413, stratify=ebay["Competitive"]
)

print("Train size:", train_df_prac.shape)
print("Test size:", test_df_prac.shape)

plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("muted")
RANDOM_STATE = 8413
```

```python
# Pre-Auction predictors (Numeric + Categorical
num_prac = ["SellerRating", "Duration", "OpenPrice"]
cat_prac = ["Category", "EndDay"]

X_train_prac = train_df_prac[num_prac + cat_prac]
y_train_prac = train_df_prac["Competitive"]
X_test_prac  = test_df_prac[num_prac + cat_prac]
y_test_prac  = test_df_prac["Competitive"]

# iii
preproc_prac = make_column_transformer(
    (StandardScaler(), num_prac),
    (OneHotEncoder(sparse_output=False, handle_unknown="ignore"), cat_prac)
)

pipe_dt_prac = make_pipeline(
    preproc_prac,
    DecisionTreeClassifier(min_samples_leaf=50, random_state=8413)
)
```

### 5.1 Hyperparameter tuning

Tuning the **max_depth** parameter to control tree complexity, using 10-fold CV with accuracy as the metric.


```python
param_grid_dt = {"decisiontreeclassifier__max_depth":[2,3,4,5,6,8,10]}
gs_dt_prac = GridSearchCV(pipe_dt_prac, param_grid_dt, scoring="accuracy", cv=10, n_jobs=-1)
gs_dt_prac.fit(X_train_prac, y_train_prac)

print("Best CV accuracy:", round(gs_dt_prac.best_score_, 3))
print("Best paramameters for this accuracy:", gs_dt_prac.best_params_)
```

### 5.2 Fitting the tuned tree 


```python
best_dt_prac = gs_dt_prac.best_estimator_

dt = best_dt_prac.named_steps["decisiontreeclassifier"]
feat_names = best_dt_prac.named_steps["columntransformer"].get_feature_names_out()

plt.figure(figsize=(18,10))
plot_tree(dt, feature_names=feat_names, class_names=["non-comp","comp"],
          filled=True, rounded=True, fontsize=9)
plt.title("Decision Tree – Practical Predictors", fontsize=16)
plt.show()
```

### 5.3 Results on the test set

We evaluate against the null baseline (≈0.54 competitive rate).


```python
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay, classification_report

y_pred = best_dt_prac.predict(X_test_prac)
acc = accuracy_score(y_test_prac, y_pred)

print("Test accuracy:", round(acc,3), " | Null ≈ 0.541")

cm = confusion_matrix(y_test_prac, y_pred, labels=[0,1])
ConfusionMatrixDisplay(cm, display_labels=["non-comp","comp"]).plot(values_format='d')
plt.title("Practical DT — Confusion Matrix"); plt.show()

print(classification_report(y_test_prac, y_pred, target_names=["non-competitive","competitive"]))
```
