### 1.2 Initial Cleanings

```python
# Rename the response column and format all other variable names
ebay = ebay_raw.rename(columns={'Competitive?': 'Competitive',
                                'sellerRating': 'SellerRating',
                                'endDay': 'EndDay'})
ebay.head()
```

```python
# convert all prices into USD (preparation)
# https://www.xe.com/currencytables/?from=GBP&date=2004-12-31
# FX on May 31, 2004
eur_to_usd = 1.2076
gbp_to_usd = 1.8080

def convert_fx(row):
    if row['Currency'] == 'EUR':
        row['ClosePrice'] *= eur_to_usd
        row['OpenPrice'] *= eur_to_usd
    elif row['Currency'] == 'GBP':
        row['ClosePrice'] *= gbp_to_usd
        row['OpenPrice'] *= gbp_to_usd
    return row
```

```python
# convert all prices into USD (continue) & get rid of the currency column
ebay = ebay.apply(convert_fx, axis = 1)
ebay['ClosePrice'] = ebay['ClosePrice'].round(2)
ebay['OpenPrice'] = ebay['OpenPrice'].round(2)

ebay = ebay.drop(columns=['Currency'])
```

```python


num_prac = ["SellerRating", "Duration", "OpenPrice"]   # quantitative, available pre-auction
cat_prac = ["Category", "EndDay"]                      # categorical, available pre-auction

X_train_prac = train_df_prac[num_prac + cat_prac]
y_train_prac = train_df_prac["Competitive"]
X_test_prac  = test_df_prac[num_prac + cat_prac]
y_test_prac  = test_df_prac["Competitive"]

# --- Pipeline: scale nums, one-hot cats, then a tuned DecisionTree ---
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import make_pipeline
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
from sklearn.model_selection import GridSearchCV

preproc_prac = make_column_transformer(
    (StandardScaler(), num_prac),
    (OneHotEncoder(sparse_output=False, handle_unknown="ignore"), cat_prac)
)

pipe_dt_prac = make_pipeline(
    preproc_prac,
    DecisionTreeClassifier(min_samples_leaf=50, random_state=8413)
)

param_grid_dt = {"decisiontreeclassifier__max_depth":[2,3,4,5,6,8,10]}
gs_dt_prac = GridSearchCV(pipe_dt_prac, param_grid_dt, scoring="accuracy", cv=10, n_jobs=-1)
gs_dt_prac.fit(X_train_prac, y_train_prac)

best_dt_prac = gs_dt_prac.best_estimator_
print("Best CV accuracy:", round(gs_dt_prac.best_score_, 3))
print("Best params:", gs_dt_prac.best_params_)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay, classification_report

y_pred = best_dt_prac.predict(X_test_prac)
acc = accuracy_score(y_test_prac, y_pred)
print("Test accuracy:", round(acc,3), " | Null  0.541")

cm = confusion_matrix(y_test_prac, y_pred, labels=[0,1])
ConfusionMatrixDisplay(cm, display_labels=[0,1]).plot(values_format='d')
plt.title("Practical DT  Confusion Matrix"); plt.show()

print(classification_report(y_test_prac, y_pred, target_names=["non-competitive","competitive"]))
```

```python
# ---------- FIX: always use the FITTED pipeline ----------
# If you used GridSearchCV:
pipe_fitted = gs_dt_prac.best_estimator_

# If you also have a separate fitted pipeline (not gridsearch), you can fallback:
# if 'pipe_dt_prac' in globals() and hasattr(pipe_dt_prac.named_steps['columntransformer'], 'transformers_'):
#     pipe_fitted = pipe_dt_prac

ct = pipe_fitted.named_steps['columntransformer']

# sanity check to avoid the same error again
assert hasattr(ct, "transformers_"), "ColumnTransformer is not fitted. Fit the pipeline before running this cell."

# ---------- get the fitted StandardScaler and its column order ----------
# build a name -> (transformer, cols) map from the fitted CT
trans_map = {name: (trans, cols) for name, trans, cols in ct.transformers_}

# your numeric transformer is named 'standardscaler' in make_column_transformer
scaler, num_cols = trans_map['standardscaler']     # scaler is fitted; num_cols in the exact order used

means  = dict(zip(num_cols, scaler.mean_))
scales = dict(zip(num_cols, scaler.scale_))

def z_to_orig(feature, z):
    return z * scales[feature] + means[feature]

# ---------- collect thresholds from the fitted decision tree ----------
from sklearn import tree

dt = pipe_fitted.named_steps['decisiontreeclassifier']  # the fitted DT inside the pipeline
feat_all = ct.get_feature_names_out()

def collect_z_thresholds(dt, feat_names, z_name):
    idx_map = {n:i for i,n in enumerate(feat_names)}
    if z_name not in idx_map: return []
    idx = idx_map[z_name]
    T = dt.tree_
    vals = []
    def walk(n):
        if T.feature[n] == idx:
            vals.append(T.threshold[n])
        if T.children_left[n] != tree._tree.TREE_LEAF:
            walk(T.children_left[n]); walk(T.children_right[n])
    walk(0)
    return sorted(set(vals))

z_v = collect_z_thresholds(dt, feat_all, "standardscaler__OpenPrice")
z_h = collect_z_thresholds(dt, feat_all, "standardscaler__SellerRating")

vlines = [z_to_orig("OpenPrice", z) for z in z_v]
hlines = [z_to_orig("SellerRating", z) for z in z_h]

# ---------- plot the scatter with exact split lines in ORIGINAL units ----------
import seaborn as sns
import matplotlib.pyplot as plt

ax = sns.scatterplot(
    data=X_test_prac.assign(competitive=y_test_prac),
    x="OpenPrice", y="SellerRating", hue="competitive",
    alpha=0.55, s=35
)
for v in vlines: plt.axvline(v, ls="--", c="k", lw=1.2)
for h in hlines: plt.axhline(h, ls="--", c="k", lw=1.2)
plt.title("Tree splits: OpenPrice (vertical)  SellerRating (horizontal)")
plt.legend(title="competitive"); plt.show()

print("OpenPrice split(s) (USD):", [round(v, 2) for v in vlines])
print("SellerRating split(s):", [round(h, 0) for h in hlines])
```
